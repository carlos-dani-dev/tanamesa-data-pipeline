```python
import os
from dotenv import load_dotenv


import pandas as pd
```

```python
load_dotenv(override=True)

survey_path = os.getenv("SURVEY_PATH")
```

```python
answ = pd.read_csv(survey_path)
```

```python
answ.columns
```

```python
columns = list(answ.columns)
columns[0] = "Submission ID"
columns[1] = "Respondent ID"
columns[2] = "Submitted at"

answ.columns = columns
```

```python
answ.columns
```

```python
print(len(answ))
print(len(answ.columns))
```

```python
import numpy as np

answ.replace(r'^\s*$', np.nan, regex=True, inplace=True)
answ = answ.dropna(how='all', axis=1)
# mantém todas as colunas que, mesmo após a remoção dos espaços, NÃO estejam em branco
answ = answ.loc[:, answ.columns.astype(str).str.strip() != ""]
# mantém todas as colunas que NÃO possuem "Unnamed"
answ = answ.loc[:, ~answ.columns.astype(str).str.contains('^Unnamed', na=False)]
```

```python
answ['index'] = range(1, len(answ) + 1)
```

```python
print(len(answ))
print(len(answ.columns))
```

```python
answ['Submitted at'] = pd.to_datetime(answ['Submitted at'])
```

```python
answ = answ[~answ['Quanto tempo você espera na fila para receber a sua refeição?'].str.contains("Sim, sou", na=False)]
answ = answ[~answ['Quanto tempo você espera na fila para receber a sua refeição?'].str.contains("Não, não sou", na=False)]
```

```python
load_dotenv()

destiny_path = os.getenv("CLEANED_DATA_PATH")
answ.to_csv(destiny_path)
```
